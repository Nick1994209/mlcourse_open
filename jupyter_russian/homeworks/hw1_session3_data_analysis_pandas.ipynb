Давайте посчитаем аналогичное значение, но на наших данных. 

Уточнения:
- Создайте новый признак ``age_years`` – возраст в годах, округлив до целых (`round`). Для данного примера отберите курящих мужчин от 60 до 64 лет включительно

- Интересуют 2 подвыборки курящих мужчин возраста от 60 до 64 лет включительно: первая с верхним артериальным давлением строго меньше 120 мм рт.ст. и концентрацией холестерина – 4 ммоль/л, а вторая – с верхним артериальным давлением от 160 (включительно) до 180 мм рт.ст. (не включительно) и концентрацией холестерина – 8 ммоль/л.

Во сколько раз (округленно, `round`) отличаются доли больных людей (согласно целевому признаку, `cardio`) в этих двух подвыборках? Посчитайте на наших данных.

**<font color='red'>Варианты:</font>**
- 2
- 3
- 4
- 5

```python
df['age_years'] = df['age'].apply(lambda x: round(x / 360))

old_smoked_mans = df[(df.gender_=='male') & (df.smoke==1) & (df.age_years>60) & (df.age_years<=64)].copy()
# cholesterol следующее: 4 ммоль/л  →→  1, 5-7 ммоль/л  →→  2, 8 ммоль/л  →→  3.
old_smoked_mans['cholesterol_'] = old_smoked_mans['cholesterol'].map({1: 4, 2: 6, 3: 8})
old_smoked_mans.head()
```

```python
first_group = old_smoked_mans[(old_smoked_mans.ap_hi < 120) & (old_smoked_mans.cholesterol_ == 4)]
second_group = old_smoked_mans[
    (old_smoked_mans.ap_hi >= 160) & (old_smoked_mans.ap_hi < 180) & (old_smoked_mans.cholesterol_ == 8)
]
```

**<font color='red'>Вопрос 6 (2 балла).</font> Постройте новый признак – BMI ([Body Mass Index](https://en.wikipedia.org/wiki/Body_mass_index)). Для этого надо вес в килограммах поделить на квадрат роста в метрах. Нормальными считаются значения BMI от 18.5 до 25. Выберите верные утверждения.**

**<font color='red'>Утверждения:</font>**
- Медианный BMI по выборке превышает норму
- У женщин в среднем BMI ниже, чем у мужчин
- У здоровых в среднем BMI выше, чем у больных
- В сегменте здоровых и непьющих мужчин в среднем BMI ближе к норме, чем в сегменте здоровых и непьющих женщин

```python
#BMI этого надо вес в килограммах поделить на квадрат роста в метрах. Нормальными считаются значения BMI от 18.5 до 25
df['BMI'] = df.apply(lambda x: x.weight / (x.height/100)**2, axis=1)
df.head()
```

```python
# Медианный BMI по выборке превышает норму (Нормальными считаются значения BMI от 18.5 до 25.)
print(df.BMI.median())
df.BMI.median() > 25
```

```python
# У женщин в среднем BMI ниже, чем у мужчин
print(df[df.gender_=='female'].BMI.mean())
print(df[df.gender_=='male'].BMI.mean())
df[df.gender_=='female'].BMI.mean() < df[df.gender_=='male'].BMI.mean()
```

```python
# У здоровых в среднем BMI выше, чем у больных
print(df[df.cardio==0].BMI.mean())
print(df[df.cardio==1].BMI.mean())
df[df.cardio==0].BMI.mean() > df[df.cardio==1].BMI.mean()
```

```python
# В сегменте здоровых и непьющих мужчин в среднем BMI ближе к норме, чем в сегменте здоровых и непьющих женщин
print(df[(df.gender_=='male') & (df.cardio==0) & (df.alco==0)].BMI.mean())
print(df[(df.gender_=='female') & (df.cardio==0) & (df.alco==0)].BMI.mean())

mean_BMI_norm = (25 + 18.5) / 2
deviation_mean_BMI_male = (df[(df.gender_=='male') & (df.cardio==0) & (df.alco==0)].BMI.mean() - mean_BMI_norm)
deviation_mean_BMI_female = (df[(df.gender_=='female') & (df.cardio==0) & (df.alco==0)].BMI.mean() - mean_BMI_norm)

# отклонение от нормы мужчин больше чем отклонение от нормы у женщин?

if deviation_mean_BMI_male > deviation_mean_BMI_female:
    print('У мужчин больше отклонение')
else:
    print('У женщин больше отклонение')
```
